```python
# dependencies
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# load the data
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training data
X_2019 = pd.get_dummies(data=train_df.drop('loan_status', axis=1))
y_2019 = pd.get_dummies(train_df['loan_status'])
X_2019
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X_2020 = test_df.drop('loan_status', axis=1)
y_2020 = pd.get_dummies(test_df['loan_status'])
y_2020
```

Hypothesis: The RandomTree model will do a better job predicting the data due to the large amount of variables. The decision tree branching will account for the data better then the logistic regression.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_2019, y_2019, random_state=1)
```

```python
#Create a training set from the 2019 loans using `pd.get_dummies()` to convert the categorical data to numeric columns. Similarly, create a testing set from the 2020 loans, also using `pd.get_dummies()`. 
# Note! There are categories in the 2019 loans that do not exist in the testing set. If you fit a model to the training set and try to score it on the testing set as is, you will get an error. You need to use code to fill in the missing categories in the testing set. 

# all non-numeric columns that were converted
home_df = pd.get_dummies(train_df['home_ownership'])
verfication_df = pd.get_dummies(train_df['verification_status'])
loan_df = pd.get_dummies(train_df['loan_status'])
initList_df = pd.get_dummies(train_df['initial_list_status'])
app_type_df = pd.get_dummies(train_df['application_type'])
hardship_df = pd.get_dummies(train_df['hardship_flag'])
debt_settlement_df = pd.get_dummies(train_df['debt_settlement_flag'])


# checked categories here and there were no such discrepancies between the datasets
# long enough to be displayed in text file; checked to ensure similarity
test_df.columns.tolist()
train_df.columns.tolist()
print("Test df column length " + str(len(test_df.columns)) + " and Training df column length " + str(len(train_df.columns)))



```

```python
# need to convert y dimensions
y_train= y_train.iloc[:,-1:].values.ravel()
y_test= y_test.iloc[:,-1:].values.ravel()
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()

classifier.fit(X_train, y_train)
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)
print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

New hypothesis: With the data being scaled, the Logistic Regression will improve as the numbers may fit with its curve better then it was unscaled. For this reason, it may surpass the RandomForest model.

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier.fit(X_train_scaled, y_train)
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

After scaling: Logistical regression became about 10% more accurate but RandomForest is still a better model and has a better testing score overall. RandomForest is therefore better at both unscaled and scaled data. It's downside is its processing time compared to Logistical regression which is much quicker to execute. However with real world data, the trend is rarely in line with Logistical Regression as things happen more randomly and in a more complex manner then it can account for. RandomForests branches can account for human decisions with its branches much more clearly.
